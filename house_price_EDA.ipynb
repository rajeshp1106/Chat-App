```python
#The attributes with high number of missing values can be dropped
df.drop(['Plot Area','Dimensions','Super Area','Society','Index','Title','Description'],axis=1,inplace=True)
```

```python
df.drop(['Transaction'],axis=1,inplace=True)
```

```python
df['Carpet Area']=df['Carpet Area'].fillna('')
#Fill empty string in place of NaN
```

```python
df['Carpet Area']=df['Carpet Area'].str.replace(' sqft','')
#Replaces 'sqft' in place of empty string
```

```python
df.tail()
```

```python
df.drop(['Status'],axis=1,inplace=True)
```

```python
categorical_columns = ['location','Furnishing','facing','overlooking','Car Parking','Ownership']
```

```python
for col in categorical_columns:
    mode_value=df[col].mode()[0]
    df[col].fillna(mode_value,inplace=True)
#Fills all empty spaces with mode value(most repeteted value) of respective attribute
```

```python
df.head()
```

```python
df.dropna(axis=0,inplace=True)
```

```python
df.isnull().sum()
```

```python
df.dtypes # Display data type of each attributes
```

```python
#Let's convert amount into crores rather than having two measures
def lac_to_cr(amount):
    if 'Lac' in amount:
        return float(amount.replace(' Lac',''))/100
    elif 'Cr' in amount:
        return float(amount.replace(' Cr',''))
```

```python
df['Amount(in rupees)']=df['Amount(in rupees)'].apply(lac_to_cr)
```

```python
df.head(20)
```

```python
df.dtypes
```

```python
df = df[~df['Carpet Area'].str.contains('[a-zA-Z]')]
#Removes the entries that contain alphabets and retrieve only numeric data
```

```python
df['Carpet Area'] = pd.to_numeric(df['Carpet Area'], errors='coerce')
#Convert to numeric values
```

```python
column_mean = df['Carpet Area'].mean(skipna=True)
column_mean
```

```python
df['Carpet Area'].fillna(column_mean, inplace=True)
```

```python
df['Balcony'] = pd.to_numeric(df['Balcony'], errors='coerce')
df = df[df['Balcony'] != '>10']
df['Balcony'] = df['Balcony'].astype(float)
```

```python
df['Bathroom'] = pd.to_numeric(df['Bathroom'], errors='coerce')
df = df[df['Bathroom'] != '>10']
df['Bathroom'] = df['Bathroom'].astype(float)
```

```python
df.head(50)
```

```python
df['Bathroom'] = df['Bathroom'].astype(float)
df['Floor'] = df['Floor'].str.replace('Ground', '0')
df['Floor'] = df['Floor'].str.replace('Upper Basement', '-1')
df['Floor'] = df['Floor'].str.replace('Lower Basement', '-2')
```

```python
df[['Empty Floor', 'Total Floor']] = df['Floor'].str.split(' out of ', expand=True)
#Seperates the floor into 2 columns with delimiter 'out of'
```

```python
df[['Empty Floor', 'Total Floor']] = df[['Empty Floor', 'Total Floor']].apply(pd.to_numeric)
```

```python
df.head()
```

```python
numeric_columns = ['Amount(in rupees)','Price (in rupees)','Carpet Area','Bathroom','Balcony','Empty Floor','Total Floor']
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
df[numeric_columns] = scaler.fit_transform(df[numeric_columns])
df[numeric_columns]
```

```python
df.drop('Floor',axis=1,inplace=True)
```
